```python
import pandas as pd
import numpy as np

```

```python
salary = pd.read_csv('/Users/beatrizrenault/Documents/IH_DATA/WEEK 7/Hackaton/ilc_di03_1_Data.csv')
#countries = pd.read_csv('countries of the world.csv')
salary.drop(['INDIC_IL','UNIT','Flag and Footnotes'], axis=1,inplace = True)
```

```python
salary_2 = salary.loc[(salary.AGE != 'Total') & (salary.SEX != 'Total')]
salary_2.dtypes
salary_2.Value = salary_2.Value.str.replace(',','')
salary_2.Value.replace({':': None},inplace =True)
salary_2
```

```python
salary_2.loc[(salary_2.GEO == 'Turkey') & (salary_2.TIME == 2018), 'Value'] = list(salary_2.loc[(salary_2.GEO == 'Turkey') & (salary_2.TIME == 2017), 'Value'])
salary_2.loc[(salary_2.GEO == 'Montenegro') & (salary_2.TIME == 2018), 'Value'] = list(salary_2.loc[(salary_2.GEO == 'Montenegro') & (salary_2.TIME == 2017), 'Value'])
salary_2.loc[(salary_2.GEO == 'Iceland') & (salary_2.TIME == 2018), 'Value'] = list(salary_2.loc[(salary_2.GEO == 'Iceland') & (salary_2.TIME == 2016), 'Value'])
```

```python
salary_3 = salary_2.loc[salary_2.TIME == 2018, :]
salary_3.Value = salary_3.Value.astype(int)
salary_3.dtypes
```

```python
salary_3 = salary_3.loc[salary_3.GEO !='European Union - 28 countries (2013-2020)']
salary_3.loc[salary_3.GEO =='Germany (until 1990 former territory of the FRG)', 'GEO'] = 'Germany'
salary_3.drop('TIME', axis=1, inplace = True)
```

```python
salary_3.columns = ['Country', 'Age Group', 'Gender', 'Yearly Wage(EUR)']
```
